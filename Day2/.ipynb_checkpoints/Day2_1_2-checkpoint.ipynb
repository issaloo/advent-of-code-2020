```python
#Load Data
with open('Day2.txt') as f:
    pass_list = f.read().splitlines()
pass_list
```

```python
def passPolicy(pass_line):
    #Split into Sections
    pass_split = pass_line.split()

    #Get Range
    range_list = [int(num) for num in pass_split[0].split('-')]

    #Get Letter
    letter = pass_split[1][0]

    #Get Password
    password = pass_split[2]

    counts = password.count(letter)

    if (counts >= range_list[0]) & (counts <= range_list[1]):
        return 1
    else:
        return 0
    
sum([passPolicy(pass_line) for pass_line in pass_list])
```

```python
def passPolicy2(pass_line):
    #Split into Sections
    pass_split = pass_line.split()

    #Get Range
    range_list = [int(num) for num in pass_split[0].split('-')]

    #Get Letter
    letter = pass_split[1][0]

    #Get Password
    password = pass_split[2]

    if (password[range_list[0]-1]==letter) != (password[range_list[1]-1]==letter):
        return 1
    else:
        return 0
sum([passPolicy2(pass_line) for pass_line in pass_list])
```
